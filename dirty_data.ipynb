```python
# Import Dependencies
import hvplot.pandas
import pandas as pd
import requests
import numpy as np
import folium
import gmaps
import datetime as dt
```

```python
canna_data = pd.read_csv("Cannabis_Legal_Status_2020.csv")
```

```python
canna_data_df["Population"] = canna_data_df["Population"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Violent Crime"] = canna_data_df["Violent Crime"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Rape2"] = canna_data_df["Rape2"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Robbery"] = canna_data_df["Robbery"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Aggravated assault"] = canna_data_df["Aggravated assault"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Property Crime Total"] = canna_data_df["Property Crime Total"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Burglary"] = canna_data_df["Burglary"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Larceny-theft"] = canna_data_df["Larceny-theft"].str.replace(",", "").astype(float)
```

```python
canna_data_df["Motor vehicle theft"] = canna_data_df["Motor vehicle theft"].str.replace(",", "").astype(float)
```

```python
# Per capita calculations
canna_data_df["Violent Crime per Capita"] = canna_data_df["Violent Crime"]/canna_data_df["Population"]
canna_data_df["Rape per Capita"] = canna_data_df["Rape2"]/canna_data_df["Population"]
canna_data_df["Robbery per Capita"] = canna_data_df["Robbery"]/canna_data_df["Population"]
canna_data_df["Aggravated Assault per Capita"] = canna_data_df["Aggravated assault"]/canna_data_df["Population"]
canna_data_df["Property Crime Total per Capita"] = canna_data_df["Property Crime Total"]/canna_data_df["Population"]
canna_data_df["Burglary per Capita"] = canna_data_df["Burglary"]/canna_data_df["Population"]
canna_data_df["Larceny-Theft per Capita"] = canna_data_df["Larceny-theft"]/canna_data_df["Population"]
canna_data_df["Motor Vehicle Theft per Capita"] = canna_data_df["Motor vehicle theft"]/canna_data_df["Population"]
```

```python
canna_data_df.head()
```

```python
State= canna_data_df["state"]
LegalStatus = canna_data_df["2020 legal status"]
LegalStatusCode = canna_data_df["2020 Legal Status Code"]
City = canna_data_df["City"]
Population = canna_data_df["Population"]
vcpc = canna_data_df["Violent Crime per Capita"]
rapc = canna_data_df["Rape per Capita"]
ropc = canna_data_df["Robbery per Capita"]
aapc = canna_data_df["Aggravated Assault per Capita"]
pctpc = canna_data_df["Property Crime Total per Capita"]
bpc = canna_data_df["Burglary per Capita"]
ltpc = canna_data_df["Larceny-Theft per Capita"]
mvtpc = canna_data_df["Motor Vehicle Theft per Capita"]




clean_data_df = pd.DataFrame({
    "City": City,
    "Legal Status": LegalStatus,
    "Legal Status Code": LegalStatusCode,
    "Population": Population,
    "Violent Crime per Capita": vcpc,
    "Rape per Capita": rapc,
    "Robbery per Capita": ropc,
    "Aggravated Assault per Capita": aapc,
    "Property Crime Total per Capita": pctpc,
    "Burglary per Capita": bpc,
    "Larceny-Theft per Capita": ltpc,
    "Motor Vehicle Theft per Capita": mvtpc
})
```
